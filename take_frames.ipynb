```python
import cv2
import os
```

```python
cam = cv2.VideoCapture('./10sec_video.mp4')

try:
    if not os.path.exists('frames'):
        os.makedirs('frames')
except OSError:
    print('Error creating directory')

count = 0

while True:
    ret, frame = cam.read()

    if ret:
        name = './frames/frame_'+str(count)+'.jpg'
        print('Creating...'+name)

        cv2.imwrite(name, frame)

        count += 1
    else:
        break
cam.release()
cv2.destroyAllWindows()
```
